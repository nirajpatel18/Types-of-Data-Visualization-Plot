# Basic area plot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#creating data
x = [2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]
y = [1.7, 1.5, 1.4, 1.8, 1.6, 1.3, 2.2, 2.4, 2.8, 3.1, 3.6]
```

```python
#area plot
plt.fill_between(x, y, color='skyblue' )
plt.xlabel('Years')
plt.ylabel('Number of units sold (in millon)')
plt.title('Yearly sales of iPhones')
plt.show()
```

# Stacked Area plot

```python
week = [1,2,3,4,5]

shift1 = [40,34,26,22,20]
shift2 = [40,32,40,45,50]
shift3 = [40,39,45,42,40]
```

```python
plt.stackplot(week, shift1, shift2, shift3, labels=['Shift 1','Shift 2', 'Shift 3'])
plt.legend()
```

# Overlapped Area Plot

```python
tempData = { 'City1':[100, 106, 112, 80],
              'City2':[80, 77, 86, 93] }
Seasons = ('Spring', 'Summer', 'Fall', 'Winter')
```

```python
df = pd.DataFrame(tempData, index = Seasons)
```

```python
df.plot(kind='area', stacked = False)
plt.ylabel('temperature in Fahrenheit')
```
